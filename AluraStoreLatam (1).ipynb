#1. Análisis de facturación



```python
# Uniremos los 4 dataframe en tiendas
# primero crear columna que indique el numero de tienda
tienda['Tienda'] = 1
tienda2['Tienda'] = 2
tienda3['Tienda'] = 3
tienda4['Tienda'] = 4

tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
tiendas.head()
```

```python
# creamos un gráfico para evaluar qué tienda tiene más ingresos
# agrupamos
tienda_ingresos = tiendas.groupby('Tienda')['Precio'].sum().reset_index()
tienda_ingresos
```

```python
#graficamos en grafico de barras
plt.figure(figsize=(8, 5))
sns.barplot(data=tienda_ingresos, x='Tienda', hue= 'Tienda',y='Precio', palette='Set2', legend=False)

#Agregar etiquetas encima de las barras
for index, row in tienda_ingresos.iterrows():
    plt.text(index, row['Precio'] + row['Precio'] * 0.01, f'{row["Precio"]:.2f}',
             ha='center', va='bottom')

plt.title('Ingresos por Tienda')
plt.ylabel('Ingresos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Ventas por categoría por tienda
fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # 2 filas, 2 columnas

# Tienda 1
cat_counts = tienda['Categoría del Producto'].value_counts().reset_index()
cat_counts.columns = ['Categoría', 'Cantidad']
sns.barplot(x='Cantidad', y='Categoría', data=cat_counts, hue='Categoría', palette='Blues', ax=axs[0, 0], legend=False)
axs[0, 0].set_title('Categorías Tienda 1')

# Tienda 2
cat_counts = tienda2['Categoría del Producto'].value_counts().reset_index()
cat_counts.columns = ['Categoría', 'Cantidad']
sns.barplot(x='Cantidad', y='Categoría', data=cat_counts, hue='Categoría', palette='Greens', ax=axs[0, 1], legend=False)
axs[0, 1].set_title('Categorías Tienda 2')

# Tienda 3
cat_counts = tienda3['Categoría del Producto'].value_counts().reset_index()
cat_counts.columns = ['Categoría', 'Cantidad']
sns.barplot(x='Cantidad', y='Categoría', data=cat_counts, hue='Categoría', palette='Oranges', ax=axs[1, 0], legend=False)
axs[1, 0].set_title('Categorías Tienda 3')

# Tienda 4
cat_counts = tienda4['Categoría del Producto'].value_counts().reset_index()
cat_counts.columns = ['Categoría', 'Cantidad']
sns.barplot(x='Cantidad', y='Categoría', data=cat_counts, hue='Categoría', palette='Purples', ax=axs[1, 1], legend=False)
axs[1, 1].set_title('Categorías Tienda 4')

# Ajustes finales
for ax in axs.flat:
    ax.set_xlabel('Cantidad')
    ax.set_ylabel('Categoría')

plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python

def plot_top_bottom(df, ax, titulo):
    # Productos más vendidos (top 5)
    top = df['Producto'].value_counts().head(5).reset_index()
    top.columns = ['Producto', 'Cantidad']
    top['Tipo'] = 'Más vendidos'

    # Productos menos vendidos (bottom 5)
    bottom = df['Producto'].value_counts().tail(5).reset_index()
    bottom.columns = ['Producto', 'Cantidad']
    bottom['Tipo'] = 'Menos vendidos'

    # Unir ambos
    combined = pd.concat([top, bottom])

    sns.barplot(data=combined, x='Cantidad', y='Producto', hue='Tipo', ax=ax)
    ax.set_title(titulo)
    ax.set_xlabel('Cantidad')
    ax.set_ylabel('Producto')

# Crear figura
fig, axs = plt.subplots(2, 2, figsize=(16, 10))

plot_top_bottom(tienda, axs[0, 0], 'Tienda 1')
plot_top_bottom(tienda2, axs[0, 1], 'Tienda 2')
plot_top_bottom(tienda3, axs[1, 0], 'Tienda 3')
plot_top_bottom(tienda4, axs[1, 1], 'Tienda 4')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular promedio del costo de envío por tienda
promedio_envio = tiendas.groupby('Tienda')['Costo de envío'].mean().round(2)
```

```python
# Gráfico de líneas del costo promedio de envío por tienda
plt.figure(figsize=(8, 5))
plt.plot(promedio_envio.index.astype(str), promedio_envio.values, marker='o', linestyle='-', color='teal')
plt.title('Promedio del Costo de Envío por Tienda (Línea)')
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio de Envío')

# Etiquetas en cada punto
for i, val in enumerate(promedio_envio):
    plt.text(i, val, f'{val:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Gráficos Combinados

```python
# Gráfico de los Ingresos y Cantidad de Productos Vendidos por Tienda
# Calcular ingresos
tiendas['Ingreso'] = tiendas['Precio']

# Agrupar por tienda
ingresos_por_tienda = tiendas.groupby('Tienda')['Ingreso'].sum()
productos_vendidos = tiendas['Tienda'].value_counts().sort_index()

# Colores personalizados
colores = [sns.color_palette('Blues')[3],
           sns.color_palette('Greens')[3],
           sns.color_palette('Oranges')[3],
           sns.color_palette('Purples')[3]]

# Crear gráfico combinado
fig, ax1 = plt.subplots(figsize=(10, 6))

# Gráfico de barras: ingresos con colores personalizados
bars = ax1.bar(ingresos_por_tienda.index.astype(str), ingresos_por_tienda, color=colores)
ax1.set_ylabel('Ingresos')
ax1.set_xlabel('Tienda')
ax1.set_title('Ingresos y Cantidad de Productos Vendidos por Tienda')

# Anotar ingresos sobre cada barra
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:,.2f}', ha='center', va='bottom', fontsize=9)

# Gráfico de línea: productos vendidos
ax2 = ax1.twinx()
ax2.plot(productos_vendidos.index.astype(str), productos_vendidos.values, color='black', marker='o', linewidth=2)
ax2.set_ylabel('Cantidad de Productos Vendidos')

plt.tight_layout()
plt.show()

```

```python
# Gráfico de los Ingresos por Tienda con su Calificación Promedio
# Asegurarse de tener las columnas necesarias
tiendas['Ingreso'] = tiendas['Precio']

# Agrupaciones
ingresos = tiendas.groupby('Tienda')['Ingreso'].sum()
calificaciones = tiendas.groupby('Tienda')['Calificación'].mean()

# Colores personalizados
colores = [sns.color_palette('Blues')[3],
           sns.color_palette('Greens')[3],
           sns.color_palette('Oranges')[3],
           sns.color_palette('Purples')[3]]

# Crear gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(ingresos.index.astype(str), ingresos.values, color=colores)
plt.title('Ingresos por Tienda con Calificación Promedio')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')

# Etiquetas con calificación promedio
for i, bar in enumerate(bars):
    altura = bar.get_height()
    calif = calificaciones.iloc[i]
    plt.text(bar.get_x() + bar.get_width()/2, altura, f'★ {calif:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

```

```python
# Gráfico de Categorías de Productos Más Vendidos por Tienda
# Contar productos por categoría y tienda
categorias_por_tienda = tiendas.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad')

# Pivotear para formato de gráfico agrupado
pivot = categorias_por_tienda.pivot(index='Categoría del Producto', columns='Tienda', values='Cantidad').fillna(0)

# Ordenar por total de productos vendidos en todas las tiendas
pivot['Total'] = pivot.sum(axis=1)
pivot = pivot.sort_values(by='Total', ascending=False).drop(columns='Total')

# Crear gráfico
pivot.plot(kind='bar', figsize=(14, 6), colormap='tab10')
plt.title('Categorías de Productos Más Vendidos por Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.legend(title='Tienda')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Mismo gráfico, pero en gráfico circular o pastel

# Para cada tienda, hacer gráfico de pastel
tiendas_unicas = tiendas['Tienda'].unique()
fig, axs = plt.subplots(1, len(tiendas_unicas), figsize=(6 * len(tiendas_unicas), 6))

for i, tienda in enumerate(tiendas_unicas):
    # Filtrar por tienda
    datos_tienda = tiendas[tiendas['Tienda'] == tienda]
    conteo = datos_tienda['Categoría del Producto'].value_counts()

    # Graficar pastel
    axs[i].pie(conteo, labels=conteo.index, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f'Categorías más vendidas Tienda - {tienda}')

plt.tight_layout()
plt.show()
```
